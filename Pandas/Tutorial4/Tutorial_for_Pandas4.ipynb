## **Creating a Pivot Table**

Columns - Name of teams

Rows - 1 to 20 overs

```python
mask = delivery['batsman_runs'] == 6
six = delivery[mask]
six.shape
```

```python
pt = six.pivot_table(index='over', columns='batting_team',values='batsman_runs',aggfunc='count')
```

```python
#Porting heat map of the above pivot table [ SEABORN IS THE REQUIRED LIBRARY ]
import seaborn as sns
```

```python
sns.heatmap(pt)
```
