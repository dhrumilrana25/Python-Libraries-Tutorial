Q1) To find the Top 5 batsman in the IPL?

->We have to sum all the runs scored by individual batsman

->We also have to sout out batsman individually

```python
#Importing all required libraries
import numpy as np
import pandas as pd
```

```python
#Loading datset into the notebool
delivery = pd.read_csv('deliveries.csv')
```

```python
#Printing top 5 Rows
delivery
```

```python
#We will make a new variable named runns and apply groupby function
runs = delivery.groupby('batsman')
```

```python
#This particular code returns sum of the runs scored by the Batsman in a Sorted Name
runs['batsman_runs'].sum().sort_values(ascending = False).head()
```

Q2) To find the Top 5 batsman who scored highest 6's in the IPL?

->We have to sum all the 6's boundries scored by individual batsman

->We also have to sout out batsman individually

```python
mask = delivery['batsman_runs'] == 6 #change 6 with 4 to find top 5 players who scored 4's
new_delivery = delivery[mask]
```

```python
new_delivery.shape[0] #17033 total 6's in IPL
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending = False)
```
