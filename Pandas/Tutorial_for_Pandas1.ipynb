```python
'''The moment we get a dataset, some of the columns and rows are not relevent so we fetch them to decrease
the crowdness of dataset so fetching (retriving) them is the best way to clean relevent data '''

data['winner'] #fetching only one column
```

```python
#To fetch multiple columns
data[['team1', 'team2', 'winner']]
```

```python
#Filtering Dataframes on a condition!
mask = data['city']=='Hyderabad' #RETURNS TRUE/FALSE
data[mask] #Returns only those values which contains cities values as 'Hyderabad'
```

doing same method of filtering but using function

```python
def getCity(city):
  mask = data['city']==city
  return data[mask]
```

```python
getCity('Rajkot') #Enter any city name it will return all the rows which contains ('City') name
```

```python
#To find out the frequency or value of counts on dataset
data['winner'].value_counts()
```

## **The Pandas Plot() functions!!**

```python
#Importing matplotlib library for plotting graphs
import matplotlib.pyplot as plt
```

```python
#Plotting Bar Chart ( require two inputs )
data['winner'].value_counts().plot(kind='bar') # y-axis counts/frequency, x-axis team name
```

```python
#Plotting Bar Chart in horizontal way
data['winner'].value_counts().plot(kind='barh') # h is added in bar
```

```python
#Plotting pie chart
data['toss_decision'].value_counts().plot(kind='pie')
```

```python
#Plotting histogram for numerical values
data['win_by_runs'].plot(kind='hist')
```

```python
#Sorting Values in the dataset [BY DEFAULT ASCENDING VALUES]
data['team2'].sort_values()
```

```python
#Sorting Values in Descending order in the dataset
data['team2'].sort_values(ascending = False)
```

```python
#Sorting values of particular column in ascending order
data.sort_values('city')
```

```python
#Sorting values of particular column in descending order
data.sort_values('city', ascending = False)
```

```python
#If you want to sort multiple columns at a time
data.sort_values(['city','date'])
```

```python
'''If you're working on a dataset there are many duplicate data, pandas have a function to remove those
for example we want to check in the column (city) it checks each row '''
data.drop_duplicates(subset=['city'])
```

```python
#Drop duplicates over multiple columns [ WORKS AS AND OPERATIONS BOTH OF COLUMNS MUST CONTAIN DUPLICATES ]
data.drop_duplicates(subset=['city','season'])
```

```python
data.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season')
```
