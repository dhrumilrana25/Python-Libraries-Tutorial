```python
#Importing required libraries
import numpy as np
import pandas as pd
```

```python
'''About Dataset

Description

Company Name - Name of the Fortune 500 company
Number of Employees - Total number of employees in the company
Previous Rank - Rank for the year 2016
Revenues - Revenue of the company for the year 2016-17 in $ millions.
Revenue Change - Percentage of Revenue change from last year
Profits - Profits of the company in $ million
Profit Change - Change in the percentage of profit from previous year.
Assets - Value of assets in $ millions
Market Value - Market Value of the company as of 3/31/17

Acknowledgements

This dataset is compiled by Fortune Magazine (www.fortune.com/fortune500).
This is the limited version of the original data'''
```

```python
#Loading dataset into the notebook name as company
company = pd.read_csv('India50.csv', encoding='ISO-8859-1')
```

```python
#To print top 5 rows of the dataset or we can say head of the dataset. [BY DEFAULT  IT SHOWS TOP 5 VALUES ]
#If you want to print top x values or row simply (x) in the bracket
company.head()
```

```python
#Grouby function can allow analysis of data by making groups
Industry = company.groupby('Industry')
```
